```python
#made func
def is_alpha(a):
    c=True
    b='qwertyuioplkjhgfdsazxcvbnm'
    for x in a:
        if x not in b:
            c=False
            break
    return(c)


def is_digit(a):
    c=True
    b=['1','2','3','4','5','6','7','8','9','0']
    for x in a:
        if x not in b:
            c=False
            break
    return c

def my_split(a):
    w=[]
    b=''
    for i in a:
        if i!=' ':
            b+=i
        else:
            if b:
                w.append(b)
                b=''
    if b:
        w.append(b)
    return w


def my_filter(f,x):
    r=[]
    for i in x:
        if f(i):
            r.append(i)
    return r

my_map= lambda f , i: [f(i[0])]+my_map(f,i[1:]) if i else []

def is_upper(a):
    if 'A'<=a<='Z':
        return True
    return False

def my_sort(a):
    n=len(a)
    for i in range(n):
        for j in range(0,n-i-1):
            if a[j]>a[j+1]:
                a[j],a[j+1]=a[j+1],a[j]
    return a

def my_sort_len_des(a):
    n=len(a)
    for i in range(n):
        for j in range(0,n-i-1):
            if len(a[j])<len(a[j+1]):
                a[j],a[j+1]=a[j+1],a[j]
    return a

def is_vowel(a):
    b='eyuioau'
    a=a.lower()
    for x  in a:
        if x in b:
            return True
    return False

def is_consonant(a):
    b='eyuioau'
    a=a.lower()
    for x  in a:
        if x in b:
            return False
    return True
```

```python
# LISTS and STRINGS 1
def analyze_text(text):
    a='aeyuio'
    u=''
    b=''
    for c in text.lower():
        if is_alpha(c) or c==' ':
            b+=c
        else:
            b+=' '
    c=[]
    d=''
    for x in b:
        if x!=' ':
            d+=x
        else:
            c.append(d)
            d=''

    if d!='':
        c.append(d)
    for x in b:
        if x in a and x not in u:
            u+=x
    result=[]
    r=''
    for x in c:
        if len(x)>=5 and x[0]==x[-1] and x not in r:
            result.append(x)
            r+=x+' '
    return (len(u), ' '.join(result))

text='Rear , apple , level, street, wertw , sdfgskhjks'
print(analyze_text(text))
```

```python
#LISTS and STRINGS 3
def is_prepinanya(a):
    c=True
    b="()!:;.,[]{}?"
    for x in a:
        if x not in b:
            c=False
            break
    return c
def top_k_words(text, k):
    text = text.lower()
    c=""
    for x in text:
        if  not is_prepinanya(x):
            c+=x
    w=my_split(c)
    b={}
    for x in w:
        if x in b:
            b[x]+=1
        else:
            b[x]=1
    a=[]
    for x,y in b.items():
        a.append((x,y))
    for i in range(len(a)):
        for j in range(0,len(a)-i-1):
            w_a , c_a=a[j]
            w_b,c_b=a[j+1]
            if c_a<c_b or (c_a==c_b and w_a>w_b):
                a[j],a[j+1]=a[j+1],a[j]
    g=[]
    for x in range(len(a)):
        if x<k:
            g.append(a[x][0])
    return g

text='I love all of you ! I can cook apple pie . i love apple .'
print(top_k_words(text, 3))
```

```python
#LISTS and STRINGS 5
def compress_text(text):
    a=text.lower()
    b={}
    for x in a:
        if x in b:
            b[x]+=1
        else:
            b[x]=1
    r=''
    n=0
    for i,j in b.items():
        if i== a[n] and j!=1:
            r+=text[n]+str(j)
            n+=j
        elif  i== a[n] and j==1:
            r+=text[n]
            n+=j
    return r

print(compress_text('dAaaaaBbbbbbcccC'))
```

```python
#LISTS and STRINGS 7
def palim(n):
    if len(n)<=1:
        return True
    if n[0]!=n[-1]:
        return False
    return palim(n[1:-1])

def palindrome_words(text):
    text= my_filter(palim ,filter(lambda x: len(x)>=3, my_split(text)))
    l=[]
    for i in text:
        if not is_prepinanya(i):
            l.append(i)


    return my_sort(l) , my_sort_len_des(l)




text='abba hello !!...!! dfggfd aba'
print( palindrome_words(text))
```

```python
#LISTS and STRINGS 9
def alternate_case_blocks(text, n):
    index=0
    r=''
    for i in range(0, len(text),n):
        b= text[i:i+n]
        if index%2==0:
            b=b.upper()
        else:
            b=b.lower()
        r+=b
        index+=1
    return r

print(alternate_case_blocks('qwertyuikgfdwertyujhgfrewsdfrtyuioffghjsosejnn ertyujhgfASFDRTDXTFyuguyCTDCYUGTJDFUKRFU', 8))
```

```python
#LISTS and STRINGS 11
def common_unique_chars(s1, s2):
    r=''
    for i in s1:
        if i==' ' or is_digit(i):
            continue
        elif i in s2 and i not in r:
            r+=i
    return r

print(common_unique_chars('qwertyuiojhgfrdeswder66ftgyh', '6frdg'))
```

```python
#LISTS and STRINGS 13
def replace_every_nth(text, n, char):
    r=''
    c=0
    w=my_split(text)
    for i in w:
        if len(i)<3:
            c+=1
            r+=i+' '
        else:
            for x in i:
                if is_digit(x):
                    r+=x
                else:
                    c+=1
                    if c%n==0:
                        r+=char
                    else:
                        r+=x
            r+=' '
    return r[:-1]

print(replace_every_nth('4hello rty567 ojjgo d f g ', 3, 'A'))
```
