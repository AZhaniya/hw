```python
#made func
def is_alpha(a):
    c=True
    b='qwertyuioplkjhgfdsazxcvbnm'
    for x in a:
        if x not in b:
            c=False
            break
    return(c)


def is_digit(a):
    c=True
    b=['1','2','3','4','5','6','7','8','9','0']
    for x in a:
        if x not in b:
            c=False
            break
    return c

def my_split(a):
    w=[]
    b=''
    for i in a:
        if i!=' ':
            b+=i
        else:
            if b:
                w.append(b)
                b=''
    if b:
        w.append(b)
    return w


def my_filter(f,x):
    r=[]
    for i in x:
        if f(i):
            r.append(i)
    return r

my_map= lambda f , i: [f(i[0])]+my_map(f,i[1:]) if i else []

def is_upper(a):
    if 'A'<=a<='Z':
        return True
    return False


```

```python
# LISTS and STRINGS 1
def analyze_text(text):
    a='aeyuio'
    u=''
    b=''
    for c in text.lower():
        if is_alpha(c) or c==' ':
            b+=c
        else:
            b+=' '
    c=[]
    d=''
    for x in b:
        if x!=' ':
            d+=x
        else:
            c.append(d)
            d=''

    if d!='':
        c.append(d)
    for x in b:
        if x in a and x not in u:
            u+=x
    result=[]
    r=''
    for x in c:
        if len(x)>=5 and x[0]==x[-1] and x not in r:
            result.append(x)
            r+=x+' '
    return (len(u), ' '.join(result))

text='Rear , apple , level, street, wertw , sdfgskhjks'
print(analyze_text(text))
```

```python
#LISTS and STRINGS 3
def is_prepinanya(a):
    c=True
    b="()!:;.,[]{}?"
    for x in a:
        if x not in b:
            c=False
            break
    return c
def top_k_words(text, k):
    text = text.lower()
    c=""
    for x in text:
        if  not is_prepinanya(x):
            c+=x
    w=my_split(c)
    b={}
    for x in w:
        if x in b:
            b[x]+=1
        else:
            b[x]=1
    a=[]
    for x,y in b.items():
        a.append((x,y))
    for i in range(len(a)):
        for j in range(0,len(a)-i-1):
            w_a , c_a=a[j]
            w_b,c_b=a[j+1]
            if c_a<c_b or (c_a==c_b and w_a>w_b):
                a[j],a[j+1]=a[j+1],a[j]
    g=[]
    for x in range(len(a)):
        if x<k:
            g.append(a[x][0])
    return g

text='I love all of you ! I can cook apple pie . i love apple .'
print(top_k_words(text, 3))
```
