```python
my_map=lambda f, i: [f(i[0])] + my_map(f, i[1:]) if i else []
def is_alpha(a):
    c = True
    b = 'qwertyuioplkjhgfdsazxcvbnm'
    for x in a:
        if x not in b:
            c = False
            break
    return (c)
def my_filter(f, x):
    r = []
    for i in x:
        if f(i):
            r.append(i)
    return r
def my_sort(a):
    n = len(a)
    for i in range(n):
        for j in range(0, n - i - 1):
            if a[j] > a[j + 1]:
                a[j], a[j + 1] = a[j + 1], a[j]
    return a


def my_sort_len_des(a):
    n = len(a)
    for i in range(n):
        for j in range(0, n - i - 1):
            if len(a[j]) < len(a[j + 1]):
                a[j], a[j + 1] = a[j + 1], a[j]
            elif len(a[j])==len(a[j + 1]) and a[j]>a[j+1]:
                a[j], a[j + 1] = a[j + 1], a[j]
    return a
#LISTS and STRINGS 7
def is_prepi(n):
    b=True
    for c in n:
        if c in '!,.?\\/':
            b=False
    return b
def palim(n):
    if len(n) <= 1:
        return True
    if n[0] != n[-1]:
        return False
    return palim(n[1:-1])
def palindrome_words(text):
    text=filter(palim,filter(
        lambda x: len(x)>=3, text.split()))
    l=[]
    for i in text:
        if not is_prepi(i):
            l.append(i)
    return my_sort(l) , my_sort_len_des(l)

text='Good morning! Ok abba aabbbaa'
print(palindrome_words(text))
```

```python
#LISTS and STRINGS 28
def moving_average(nums, k):
    r=[]
    for i in range(len(nums)-k+1):
        b=0
        c=True
        for j in range(i,i+k):
            if nums[j]<0:
                c=False
                break
            b+=nums[j]
        if c:
            r.append(b/k)
    return r
print(moving_average([1,2,3,4,5,6,7,8,9,55], 3))
```

```python
#DICT AND SET 19
def filter_by_digit_sum(nums):
    r=set()
    for x in nums:
        if x%2!=0 and sum(int(d) for d in str(x))%2==0:
            r.add(x)
    return x
print(filter_by_digit_sum({12,34,56,77,89}))
```

```python
#DICT AND SET 26
def remove_elements_with_common_digits(s):
    r=set()
    u=[]
    c={}
    for x in s:
        x=str(x)
        for i in x:
            if i in c:
                c[i]+=1
            else:
                c[i]=1
    for i ,j in c.items():
        if j==1:
            u.append(i)
    for x in s:
        x=str(x)
        if all(i in u for i in x):
            r.add(int(x))
    return r
print(remove_elements_with_common_digits({333,444,5673,890}))
```
