```python
#made func
def is_alpha(a):
    c = True
    b = 'qwertyuioplkjhgfdsazxcvbnm'
    for x in a:
        if x not in b:
            c = False
            break
    return (c)


def is_digit(a):
    c = True
    b = ['1', '2', '3', '4', '5', '6', '7', '8', '9', '0']
    for x in a:
        if x not in b:
            c = False
            break
    return c


def my_split(a):
    w = []
    b = ''
    for i in a:
        if i != ' ':
            b += i
        else:
            if b:
                w.append(b)
                b = ''
    if b:
        w.append(b)
    return w


def my_filter(f, x):
    r = []
    for i in x:
        if f(i):
            r.append(i)
    return r


my_map = lambda f, i: [f(i[0])] + my_map(f, i[1:]) if i else []


def is_upper(a):
    if 'A' <= a <= 'Z':
        return True
    return False


def my_sort(a):
    n = len(a)
    for i in range(n):
        for j in range(0, n - i - 1):
            if a[j] > a[j + 1]:
                a[j], a[j + 1] = a[j + 1], a[j]
    return a


def my_sort_len_des(a):
    n = len(a)
    for i in range(n):
        for j in range(0, n - i - 1):
            if len(a[j]) < len(a[j + 1]):
                a[j], a[j + 1] = a[j + 1], a[j]
            elif len(a[j])==len(a[j + 1]) and a[j]>a[j+1]:
                a[j], a[j + 1] = a[j + 1], a[j]
    return a



def is_vowel(a):
    b = 'eyuioau'
    a = a.lower()
    for x in a:
        if x in b:
            return True
    return False


def is_consonant(a):
    b = 'eyuioau'
    a = a.lower()
    for x in a:
        if x in b:
            return False
    return True


def my_count(a, b):
    c = 0
    for i in a:
        if i == b:
            c += 1
    return c

def my_min(a):
    r=a[0]
    for x in a:
        if r>x:
            r=x
    return r

def my_zip(a,b):
    r=[]
    n=my_min([len(a),len(b)])
    for i in range(n):
        r.append((a[i],b[i]))
    return r

def is_simple(a):
    for i in range(2,a):
        if a%i==0:
            return False
    return True

def summa(a):
    r=0
    for i in a:
        r+=i
    return r
```

```python
# LISTS and STRINGS 1
def analyze_text(text):
    a = 'aeyuio'
    u = ''
    b = ''
    for c in text.lower():
        if is_alpha(c) or c == ' ':
            b += c
        else:
            b += ' '
    c = []
    d = ''
    for x in b:
        if x != ' ':
            d += x
        else:
            c.append(d)
            d = ''

    if d != '':
        c.append(d)
    for x in b:
        if x in a and x not in u:
            u += x
    result = []
    r = ''
    for x in c:
        if len(x) >= 5 and x[0] == x[-1] and x not in r:
            result.append(x)
            r += x + ' '
    return (len(u), ' '.join(result))


text = 'Rear , apple , level, street, wertw , sdfgskhjks'
print(analyze_text(text))
```

```python
#LISTS and STRINGS 3
def is_prepinanya(a):
    c = True
    b = "()!:;.,[]{}?"
    for x in a:
        if x not in b:
            c = False
            break
    return c


def top_k_words(text, k):
    text = text.lower()
    c = ""
    for x in text:
        if not is_prepinanya(x):
            c += x
    w = my_split(c)
    b = {}
    for x in w:
        if x in b:
            b[x] += 1
        else:
            b[x] = 1
    a = []
    for x, y in b.items():
        a.append((x, y))
    for i in range(len(a)):
        for j in range(0, len(a) - i - 1):
            w_a, c_a = a[j]
            w_b, c_b = a[j + 1]
            if c_a < c_b or (c_a == c_b and w_a > w_b):
                a[j], a[j + 1] = a[j + 1], a[j]
    g = []
    for x in range(len(a)):
        if x < k:
            g.append(a[x][0])
    return g


text = 'I love all of you ! I can cook apple pie . i love apple .'
print(top_k_words(text, 3))
```

```python
#LISTS and STRINGS 13
def replace_every_nth(text, n, char):
    r = ''
    c = 0
    w = my_split(text)
    for i in w:
        if len(i) < 3:
            c += 1
            r += i + ' '
        else:
            for x in i:
                if is_digit(x):
                    r += x
                else:
                    c += 1
                    if c % n == 0:
                        r += char
                    else:
                        r += x
            r += ' '
    return r[:-1]


print(replace_every_nth('4hello rty567 ojjgo d f g ', 3, 'A'))
```

```python
#LISTS and STRINGS 16
def transform_list(nums):
    r = []
    for i in nums:
        if i < 0:
            continue
        elif i % 2 == 0:
            r.append(i ** 2)
        elif i % 2 == 1 and i > 10:
            s = 0
            while i > 0:
                d = i % 10
                s += d
                i = i // 10
            r.append(s)
        elif i % 2 == 1:
            r.append(i)
        elif i < 0:
            pass
    return r


print(transform_list([23, -1, 34, 5, 7, 13, -45]))
```

```python
#DICT AND SET 26
def remove_elements_with_common_digits(s):
    u=[]
    r=set()
    c={}
    for x in s:
        x=str(x)
        for i in x:
            if i in c:
                c[i]+=1
            else:
                c[i]=1
    for i,j in c.items():
        if j==1:
            u.append(i)
    for x in s:
         x=str(x)
         if all(y in u for y in x):
             r.add(int(x))
    return r

s={33,2,14,67,45}
print(remove_elements_with_common_digits(s))
```

```python
#DICT AND SET 30
def partition_by_sum_parity(s):
    a=set()
    b=set()
    r=[]
    for i in s:
        t=0
        for j in str(i):
            t+=int(j)
        if t%2==0:
            a.add(i)
        else:
            b.add(i)
    r.append(a)
    r.append(b)
    return tuple(r)

s={11,32,66,78,34,189}
print(partition_by_sum_parity(s))
```

```python
#DICT AND SET 34
def top_k_smallest_unique(nums, k):
    a=set()
    r={}
    for i in nums:
        if i in r:
            r[i]+=1
        else:
            r[i]=1
    while len(a)<k and r:
        m=None
        n=None
        for i in r:
            if m is None or r[i]<m :
                m=r[i]
                n=i
        a.add(n)
        del r[n]
    return a
print(top_k_smallest_unique([1,1,1,2,2,5,5,5,5,8,9], 3))
```

```python
#DICT AND SET 36
def all_subsets_of_size_k(s, k):
    s=list(s)
    n=len(s)
    i=list(range(k))
    r=[]
    while True:
        a=set()
        for x in i:
            a.add(s[x])
        r.append(a)
        for x in range(k-1,-1,-1):
            if i[x]!=x+n-k:
                break
        else:
            break

        i[x]+=1
        for y in range(x+1,k):
            i[y]=i[y-1]+1
    return r

print(all_subsets_of_size_k({1,2,3,4},3))
```

```python
#DICT AND SET 38
def multi_symmetric_difference(sets_list):
    r=sets_list[0].copy()
    for i in sets_list[1:]:
        a=set()
        for x in r:
            if x not in i:
                a.add(x)
        for x in i:
            if x not in r:
                a.add(x)
        r=a
    return r

print(multi_symmetric_difference([{1,2,3,4,5},{3,4,5,5,6},{6,7,8}]))
```

```python
#1
def analyze_students(data):
    v=set('aeyuioAEYUIO')
    filtred_s=[]
    for s in data:
        if not any(c.isdigit() for c in s['name']):
            filtred_s.append({
                "name": s["name"].title(),
                "grades": s["grades"],
                "comments": s["comments"]
            })
    for s in filtred_s:
        a=[]
        for g in s['grades']:
            if g<=0:
                continue
            elif g>=10:
                a.append(g**2)
            elif g<10 and g%2!=0:
                a.append(sum(int(d) for d in str(g)))
            else:
                a.append(g)
        s["processed_grades"]=a
    uniq_comm=[]
    vowel=set()
    for s in filtred_s:
        comm=' '.join(s['comments'])
        comm=''.join(filter(lambda c: not is_prepinanya(c) ,comm))
        w=comm.split()
        uniiq_w=set(
            filter(lambda x: len(x)>=4
                   and x.lower()!=x[::-1].lower(),
                   w)
        )
        s['comment_words']=uniiq_w
        s["comment_vowels"]=set(
            c.lower() for w in uniiq_w for c in w if c in v
        )
        uniq_comm.append(uniiq_w)
    count_w={}
    for i in set().union(*uniq_comm):
        count=sum(1 for y in uniq_comm if i in y)
        if count>=2:
            count_w[i.lower()]=count
    count_w=dict(sorted(count_w.items(), key=lambda x:(-x[-1],x[0])))
    for s in filtred_s:
        vowel.update(s["comment_vowels"])
    s_avg=[]
    for s in filtred_s:
        avg=sum(s["processed_grades"])/len(s["processed_grades"]) \
            if s["processed_grades"] else 0
        s_avg.append((s['name'],avg))
    s_avg.sort(key=lambda x:(-x[1],x[0]))
    s_by_avg=[name for name ,_ in s_avg]
    s_len={}
    for s in filtred_s:
        l=len(s['name'])
        s_len.setdefault(l,[]).append(s['name'])
    s_len={k:list(set(v)) for k,v in s_len.items()}
    s_result=[{
        'name':s['name'],
        "processed_grades":s["processed_grades"],

    } for s in filtred_s]

    r={
        "students":s_result,
        "word_counts": count_w,
        "all_vowels": vowel,
        "students_by_avg": s_by_avg,
        "students_by_name_length": s_len
    }
    return r

data=[
    {
        "name": "Alice123",
        "grades": [12, 9, 15, 8],
        "comments": ["Good work", "excellent effort", "Needs Improvement"]
    },
    {
        'name':'Alise',
        'grades': [33,5,9,-9],
        'comments':["Good work", "excellent effort"]
    },
    {
        'name':'Bob',
        'grades':[22,56],
        'comments':['Amazing!']

    }
]
print(analyze_students(data))
```
