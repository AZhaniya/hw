```python
#2.2
def filter_words(words):
    for x in words:
        if len(x)<4:
            continue
        elif "а" in x:
            yield "с а"
        else :
            yield x
words = ["кот", "машина", "арбуз", "дом"]
for w in filter_words(words):
    print(w)
```

```python
#3.3
words = ["кот", "машина", "ананас", "дом","увперт"]
r= [x for x in words if len(x)>4 and "а" not in x ]
print(r)
```
