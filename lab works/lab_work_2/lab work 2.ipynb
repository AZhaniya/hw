```python
#1
with open('shop_logs.txt', 'r', encoding='utf-8') as f:
    lines=f.readlines()
user=set()
buys=0
summa=0
user_buy={}
for i in lines:
    a=i.strip().split(';')
    date=a[0]
    u=a[1]
    act=a[2]
    user.add(u)
    if act=='BUY':
        buys+=1
        summa+=int(a[3])
        if u in user_buy:
            user_buy[u]+=int(a[3])
        else:
            user_buy[u]=int(a[3])
top=max(user_buy, key=user_buy.get)
avarage=summa/buys

with open('report.txt','w',encoding='utf-8') as f:
    f.write(f'Уникальных пользователей: {len(user)}\n')
    f.write(f'Всего покупок:{buys}\n')
    f.write(f'Общая сумма:{summa}\n')
    f.write(f'Самый активный покупатель: {top}\n')
    f.write(f'Средний чек:{avarage}\n')
```

```python
#2
import csv
d=[{"name": "Ali", "department": 'IT', "salary": 500000},
    {"name": "Dana", "department": 'HR', "salary": 300000},
   {"name": "Arman", "department": 'IT', "salary": 600000},
   {"name": "Aruzhan", "department": 'Marketing', "salary": 400000},
   {"name": "Dias", "department": 'IT', "salary": 450000}]
with open('employees.csv', 'w', encoding='utf-8', newline='') as f:
    a=['name','department','salary']
    b= csv.DictWriter(f, fieldnames=a)
    b.writeheader()
    b.writerows(d)


worker=[]
dep_sal={}
summa=0
c=0

with open('employees.csv', 'r', encoding='utf-8', newline='') as f:
    r=csv.DictReader(f)
    for i in r:
        i['salary']=int(i['salary'])
        worker.append(i)
        summa+=int(i['salary'])
        c+=1
        d=i['department']
        if d in dep_sal:
            dep_sal[d].append(i['salary'])
        else:
            dep_sal[d]=[i['salary']]
avarage=summa/c
dep_avg={}
top_w=max(worker, key= lambda x: x['salary'])
for x , y in dep_sal.items():
    dep_avg[x]=sum(y)/len(y)
top_dep=max(dep_avg, key= dep_avg.get)
filtred= filter(lambda x: x['salary']>avarage,worker)

with open('high_salary.csv', 'w', encoding='utf-8',newline='') as f:
    filed=['name', 'department','salary']
    b= csv.DictWriter(f, fieldnames=filed)
    b.writeheader()
    b.writerows(filtred)

print(f'Средняя зарплата: {avarage}')
print('Средняя по отделам:')
for x,y in dep_avg.items():
    print(f'{x}:{y}')
print(f'Отдел с самой высокой средней зарплатой: {top_dep}')
print(f'Самый высокооплачиваемый сотрудник: {top_w['name']}')
```

```python
#3
import  json
with open('orders.json', 'r',encoding='utf-8') as f :
    r=json.load(f)
summa=0
quantity_oforder={}
quantity_item={}
top_use=''
max_total=0
for x in r:
    summa+=x['total']
    if x['user'] in quantity_oforder:
        quantity_oforder[x['user']]+=1
    else:
        quantity_oforder[x['user']]=1
    for item in x['items']:
        if item in quantity_item:
            quantity_item[item]+=1
        else:
            quantity_item[item]=1
    if x['total']>max_total:
        max_total=x['total']
        top_use=x['user']
popular_item=max(quantity_item,key =quantity_item.get)
result={
    "total_revenue": summa ,
     "top_user": top_use ,
     "most_popular_item": popular_item,
      "total_orders":  quantity_oforder
}

with open('summary.json','w', encoding='utf-8') as f:
    json.dump(result, f , indent=2, ensure_ascii=False)
```

```python
#4
import csv
import json
d = [
    {'user_id': 'user_1' , 'amount': 5000},
    {'user_id': 'user_2' , 'amount': 10000},
    {'user_id': 'user_1' , 'amount': 700000},
    {'user_id': 'user_3' , 'amount': 3000},
    {'user_id': 'user_2' , 'amount': 900000},
    {'user_id': 'user_4' , 'amount': 2000}
]
with open('transactions.csv', 'w', encoding='utf-8', newline='') as f:
    a = ['user_id' , 'amount']
    b = csv.DictWriter(f, fieldnames=a)
    b.writeheader()
    b.writerows(d)

sus_trans=0
sus_user=[]
users=set()
amount_sus=0
transaction={}
f=open('transactions.csv', 'r', encoding='utf-8', newline='')

for i in csv.DictReader(f):
    users.add(i['user_id'])
    if  i['user_id'] in transaction:
        transaction[i['user_id']]+=1
    else:
        transaction[i['user_id']]=1
    if int(i['amount'])>500000:
        amount_sus+=int(i['amount'])
        sus_trans+=1
for x, y in transaction.items():
    if y>3:
        sus_user.append(x)


with open('fraud_report.txt', 'w', encoding='utf-8') as f :
    f.write(f'Подозрительных транзакций: {sus_trans}\n')
    f.write(f'Подозрительных пользователей: {sus_user}\n')
    f.write(f'Список пользователей:{users}\n')
    f.write(f'Общая сумма подозрительных операций: {amount_sus}\n')

with open('fraud_users.json', 'w', encoding='utf-8') as f :
    json.dump(sus_user,f,indent=2, ensure_ascii=False)
```
